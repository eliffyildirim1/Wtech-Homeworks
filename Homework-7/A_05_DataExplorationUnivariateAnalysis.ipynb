**(2)** Consider a state (eg California) and plot the yearly change of total revenues and total spending. Do these two variables change by years? Do you have years of summit and bottom?

```python
import datetime as dt

new_df = df[df['STATE'].isin(['ALABAMA'])]
plt.figure(figsize=(10, 5))
ax = sns.lineplot(x = 'YEAR', y = 'TOTAL_REVENUE', data = new_df, 
             lw = 5, label='Total Revenue',
             markers=False, dashes=True)
ax = sns.lineplot(x = 'YEAR', y = 'TOTAL_EXPENDITURE', data = new_df, 
             lw = 3, label='Total Expenditure',
             markers=False, dashes=True)
plt.xticks(rotation = 45, fontsize = 9)

plt.show()
```

**(3)** In which course are students more successful in your chosen state? Mathematics or reading?

```python
import datetime as dt

plt.figure(figsize=(10, 5))
ax = sns.lineplot(x = 'YEAR', y = 'AVG_MATH_8_SCORE', data = new_df, 
             lw = 5, label='Avg Math Score',
             markers=False, dashes=True)
ax = sns.lineplot(x = 'YEAR', y = 'AVG_READING_8_SCORE', data = new_df, 
             lw = 3, label='Avg Reading Score',
             markers=False, dashes=True)
plt.xticks(rotation = 45, fontsize = 9)

plt.show()
```

**(4)** What is the distribution of mathematics and reading grades?

```python
import warnings
warnings.filterwarnings('ignore')
sns.set(font = 'Century Gothic')
title_style = {'family': 'Century Gothic', 'color': 'darkred', 'size': 20 }
axis_style  = {'family': 'Century Gothic', 'color': 'darkblue', 'size': 15 }

plt.figure(figsize=(15, 5))
sns.set(style="darkgrid")
plt.subplot(1,2,1)
sns.distplot(new_df.AVG_READING_8_SCORE.dropna(), rug=True,rug_kws={"color": "r"}, 
             kde_kws={"color": "#FF914D", "lw": 3}, color = '#FF914D', bins=10)
plt.title("Avg Reading Score", fontdict = title_style)

plt.subplot(1,2,2)
sns.distplot(new_df.AVG_MATH_8_SCORE.dropna(),  bins=10, rug=True, rug_kws={"color": "g"},
                  kde_kws={"color": "k", "lw": 3},
                  hist_kws={"histtype": "step", "linewidth": 3,
                            "alpha": 1, "color": "r"})
plt.title("Avg Math Score ", fontdict = title_style)

plt.show()
```

**(5)** You may have noticed that there are many missing values in the mathematics and reading notes. Complete the missing values with mean, median, and interpolation. Compare the impact of these techniques on the distribution of grades.

```python
import warnings
warnings.filterwarnings('ignore')

new_df["AVG_MATH_8_SCORE"].fillna(new_df["AVG_MATH_8_SCORE"].mean(), inplace=True)
new_df["AVG_READING_8_SCORE"].fillna(new_df["AVG_READING_8_SCORE"].mean(), inplace=True)

sns.set(font = 'Century Gothic')
title_style = {'family': 'Century Gothic', 'color': 'darkred', 'size': 20 }
axis_style  = {'family': 'Century Gothic', 'color': 'darkblue', 'size': 15 }

plt.figure(figsize=(15, 5))
sns.set(style="darkgrid")
plt.subplot(1,2,1)
sns.distplot(new_df.AVG_READING_8_SCORE.dropna(), rug=True,rug_kws={"color": "r"}, bins=10)
plt.title("Avg Reading Score | Mean", fontdict = title_style)

plt.subplot(1,2,2)
sns.distplot(new_df.AVG_MATH_8_SCORE.dropna(),  bins=10, rug=True, rug_kws={"color": "g"},
                  kde_kws={"color": "k", "lw": 3, "label": "KDE"},
                  hist_kws={"histtype": "step", "linewidth": 3,
                            "alpha": 1, "color": "r"})
plt.title("Avg Math Score | Mean", fontdict = title_style)

plt.show()
```
