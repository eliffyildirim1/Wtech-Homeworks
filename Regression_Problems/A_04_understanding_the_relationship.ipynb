```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.api.types import is_numeric_dtype
from sklearn import metrics
import statsmodels.api as sm
import math
from sklearn import linear_model

df = pd.read_csv("C:/Users/Elif/data/weatherHistory.csv")
```

```python

df["Temp_diff"] = df["Temperature (C)"] - df["Apparent Temperature (C)"]

Y = df["Temp_diff"]

X =df[["Humidity","Wind Speed (km/h)"]]

lrm = linear_model.LinearRegression()

# fit method estimates the coefficients using OLS
lrm.fit(X, Y)



X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

The estimated coefficients are statistically **significant**  because p-values< 0.05.

$$ TempatureDifference = -2.4381 + 3.0292Humidity + 0.1193Wind Speed (km/h) $$

```python
df=df.drop(['PoolQC','MiscFeature','Fence','Alley'], axis=1)
```

```python
def fix_missing(df, col, name):
    if is_numeric_dtype(col):
        df[name] = col.fillna(col.median())    
for n, c in df.items():
        fix_missing(df, c, n)
```

```python
Y = df["SalePrice"]

X = df[["YearBuilt","TotalBsmtSF","1stFlrSF","FullBath","GarageArea",'GrLivArea','OverallQual']]
X = sm.add_constant(X)

results = sm.OLS(Y,X).fit()
display(results.summary())
```

Features are statistically significant except FullBath. Let's build the model without the FullBath feature.

```python
X_ = df[["YearBuilt","TotalBsmtSF",
                 "1stFlrSF","GarageArea",'GrLivArea','OverallQual']]
X_ = sm.add_constant(X_)

results = sm.OLS(Y,X_).fit()
display(results.summary())
```
